```python
print(range(10))
for i in range(10):
    print(i)

```

```python
str = "Aditya"
for i in str:
    print(i)
```

```python
l = [1,2,"hello",True]
for i in l:
    print(i)
```

```python
for i in range(2):
    for j in range(10):
        print(i,j)
```

```python
for i in range(100):
    if i%2 == 0:
        print(i," even")
    else:
        print(i," odd")
```

### While Loop

```python
count = 0
while count <= 10:
    print(count)
    count+=1
    
```

```python
counter = 0
while counter < 5:
    print("Within the while loop and the iteration is", counter)
    counter += 1
print("End of while loop")

name = ""
while name != 'root':
    name = input('Enter the username:')
print("Thank you")
```

## Transfer Statements

```python
for i in range(100):
    if i == 24:
        break
    else:
        print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
# loops with else block
cart = [10, 20, 30, 50]
for item in cart:
    if item >= 100:
        print("This order can't be processed")
        break
    print(item)
else:
    print("All cart items processed in the order")
```

# Compound Interest

```python
p = 1000
r = 10
t = 5

for _ in range(t):
    p += p*r/100
print(p)
```
